```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
fig.show()
```

```python
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["tuffy","Z", "Nami", "Usopp", "Sanji"])
y2 = np.random.randint([110,180,240,99,220])
fig, (ax1, ax2) = plt.subplots(ncols=2,
                             nrows=1,
                             figsize =(7,3)

                               )

ax1.scatter(x1,y1,marker = '*', color='red')
ax2.bar(x2, y2, width = 0.5, color = 'pink', label='Bars')
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                   ['ax2', 'ax3']],
                                   figsize = (7,4)
                                   )
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2) 
someaxes['ax3'].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                   ['ax2', 'ax3']],
                                   figsize = (7,4)
                                   )
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2) 
someaxes['ax3'].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 fast 2 Furious', fontsize=18)
```

```python
x = np.array(["tuffy","Z", "Nami", "Usopp", "Sanji"])
y1 = np.random.randint([110,180,240,99,220])
y2 = np.random.randint([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))
ax.bar(x,y1,color='indigo')
ax.plot(x,y2,color='red')
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color='red')
ax.errorbar(x,y2,yerr=y2_sd, fmt='none',
            ecolor = 'indigo',
            elinewidth=4,
            capsize=6,
            capthick=4
            )
```
