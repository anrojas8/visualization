```python
import plotly.graph_objects as go
import nbformat
```

```python
x1= np.array(["luffy", "zorro", "nami", "usopp", "sanji"])
y1=np.array([110, 180, 240, 99, 220])
```

```python
graph =go.Figure()

graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(title="Pirates Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score"
                    )
```

```python
graph=go.Figure()


graph.add_trace(go.Scatter(x=x1, 
                           y=y1,
                           mode = 'markers',
                           marker = dict(size=15,
                                         color='hotpink',
                                         opacity=1,
                                         line=dict(width=5, color='purple')
                                         
                                         )
                           
                           ))
graph.update_layout(title="Interactive Pirate Plot",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score",
                    width=500,
                    height=500
                    )
```

```python
from wordcloud import WordCloud
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",on_bad_lines='skip')
```

```python
text = "".join(each for each in df.quote)
wordcloud = WordCloud(background_color='white',
                      colormap ='inferno',
                      random_state=613).generate(text)
                      
fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud)
ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn3_unweighted
```

```python
A = set(["apple", "banana", "watermelon"])
B = set(["pumpking", "blueberry", "apple", "key"])
```

```python
diagram = venn2([A,B],
                set_labels= ("Fruits", "Pies"),
                set_colors = ('blue', 'red'),
                alpha=0.5
                )
diagram.get_label_by_id("10").set_text("\n".join(A-B)) #left
diagram.get_label_by_id("11").set_text("\n".join(A&B)) #middle
diagram.get_label_by_id("01").set_text("\n".join(B-A)) #
```
